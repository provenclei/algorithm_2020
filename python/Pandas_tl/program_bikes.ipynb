```python
bikes = pd.read_csv("./data/bikes.csv", encoding="latin-1", sep=';', parse_dates=['Date'], index_col='Date')
bikes.head()
```

```python
bikes.dropna(how='all', axis=1)
```

```python
# 填充缺失值
bikes.mean(1)
bikes.T.mean(0)
```

```python
# 默认为每列进行填充
# 如果对每行进行填充则需转置
bikes.T.fillna(bikes.T.mean()).T
```

```python
bikes.index
bikes.index.weekday
bikes.loc[:,'weekday'] = bikes.index.weekday
bikes
```

```python
bikes_week = bikes.groupby('weekday').mean()
bikes_week
bikes_week.index = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
bikes_week
```

```python
type(bikes.mean(axis=1))
type(bikes.mean(axis=1).to_frame())
bikes_avg = bikes.mean(axis=1).to_frame()
bikes_avg.columns = ["bikes_avg"]
bikes_avg.head()
```

```python
bikes_avg.loc[:, "weekday"] = bikes_avg.index.weekday
bikes_avg.head()
```

```python
weekday_avg = bikes_avg.groupby("weekday").mean()
weekday_avg.index = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_avg
```

```python
bikes_avg["bikes_avg"].plot()
```
