```python
stud_alcoh = df.loc[: , "school":"guardian"]
stud_alcoh.head()
```

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False

stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh.head()
```

```python
def times10(x):
    if type(x) is int:
        return 10 * x
    return x

stud_alcoh.applymap(times10)
```
